```
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/gvsanthu10/Dicom/master/Data_Entry_2017.csv')
```

```
data = data.iloc[:,0:2]
data.tail()
```

```
data.columns = ['ID', 'labels']
```

```
data = data.apply(lambda x: x.str.replace('|',','))
```

```
data.head()
```

```
df = pd.DataFrame(data)
       

df = df['labels'].str.split('\s*,\s*', expand=True)\
                .stack().str.get_dummies().sum(level=0)
print(df)
```

```
df.head()
```

```
df=df.iloc[:,9:]
```

```
df.iloc[:,:3]
```

```
predata = pd.concat([data, df], axis=1)
predata = predata.drop(['labels'], axis=1)
```

```
predata.head()
```

```
predata=predata.drop(['Pleural_Thickening','Pneumonia', 'Pneumothorax'], axis=1)
```

```
predata.to_csv('out.csv')
```
